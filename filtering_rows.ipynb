# Filtering rows in a DataFrame

Fonte: https://towardsdatascience.com/8-ways-to-filter-pandas-dataframes-d34ba585c1b8

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'name':['Jane','John','Ashley','Mike','Emily','Jack','Catlin'],
'ctg':['A','A','C','B','B','C','B'],
'val':np.random.random(7).round(2),
'val2':np.random.randint(1,10, size=7)})
```

```python
display(df)
```

```python
# Filtrando por condição de uma coluna
df[df.val > 0.5]
```

```python
# Filtrando por condição de duas ou mais colunas
df[(df.val > 0.5) & df.name.str.startswith("Jo")]
```

```python
# Filtrando por condição de uma coluna contendo string

'''CONTAINS = contém o caractere em qualquer posição'''
df[df.name.str.contains('y')]
```

```python
'''STARTSWITH = contém o caractere na primeira posição'''
df[df.name.str.startswith('J')]
```

```python
'''~(tilde) = exclui as linhas que atendem a condição'''
df[~df.name.str.startswith('J')]
```

```python
# Query

'''Permite combinar condições entre aspas simples'''
df.query('ctg == "B" and val > 0.3')
```

```python
df.iloc[3:5, :] #rows 3 and 4, all columns
```

```python
df.loc[3:5, :] #rows 3 to 5, included, all columns
```
